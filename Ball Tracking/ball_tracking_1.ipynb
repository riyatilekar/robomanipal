```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
cap = cv2.VideoCapture("ball.mp4")

fps = 30

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

fourcc = cv2.VideoWriter_fourcc(*'XVID')

out = cv2.VideoWriter(
    'tracked_ball.avi',
    fourcc,
    fps,
    (width, height)
)

lower_green = np.array([20, 80, 80])
upper_green = np.array([45, 255, 255])
```

```python
while True:

    ret, frame = cap.read()

    if not ret:
        break

    hsv = cv2.cvtColor(
        frame,
        cv2.COLOR_BGR2HSV
    )

    mask = cv2.inRange(
        hsv,
        lower_green,
        upper_green
    )

    contours, _ = cv2.findContours(
        mask,
        cv2.RETR_EXTERNAL,
        cv2.CHAIN_APPROX_SIMPLE
    )

    if len(contours) > 0:

        largest = max(
            contours,
            key=cv2.contourArea
        )

        (x, y), radius = cv2.minEnclosingCircle(
            largest
        )

        if radius > 5:

            cv2.circle(
                frame,
                (int(x), int(y)),
                int(radius),
                (0, 255, 0),
                3
            )

    out.write(frame)

cap.release()
out.release()

print("Done")
```
