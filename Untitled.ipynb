```python
# Importing all the libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# REading csv file
df = pd.read_csv("weatherAUS.csv")
df.head()
```

```python
# this shows the total number of missing values in each column

df.isnull().sum()
```

```python
# this is to calculate the percentage of missing values in each column

missing_percent = df.isnull().mean() * 100
print(missing_percent.sort_values(ascending=False))

# here the columns Sunshine, Evaporation, Cloud3pm, Cloud9am gave more than 40% missing values so drop em
```

```python
df.drop(['Sunshine', 'Evaporation', 'Cloud3pm', 'Cloud9am'], axis=1, inplace=True)
```

```python
# For handling the missing data using fillna
# For numeric features, replace the missing values with median as it is the middle value and outliers do not affect it

# For categories, replace it with mode
# Since categories are yes and no, replace them with the mode which is the most occuring one out the 2

for col in df.select_dtypes(include=['float64', 'int64']):
    df[col].fillna(df[col].median(), inplace=True)

for col in df.select_dtypes(include=['object']):
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# Now to check if the missing values have been handled
df.isnull().sum()
# it worked 
```

```python
# Converting Date to datetime object
# Date is a string and cannot be used 

df['Date'] = pd.to_datetime(df['Date'])

df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day

df.drop('Date', axis=1, inplace=True)
```

```python
df[['year','month','day']].head()
```

```python
# Encoding strings to numerical values
df = pd.get_dummies(df, drop_first=True)

# Check result
print(df.info())
# it shows that all the datatypes are float and boolean 
```

```python
# handling extreme values
# using IQR method and remove rows beyond the range
cols = ['Rainfall', 'WindSpeed9am', 'WindSpeed3pm', 'Temp3pm']

for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df = df[(df[col] >= lower) & (df[col] <= upper)]
```

```python
df.shape
```

```python
# Standardisation(z score)
num_cols = df.select_dtypes(include=['float64', 'int64']).columns

for col in num_cols:
    mean = df[col].mean()
    std = df[col].std()
    
    df[col] = (df[col] - mean) / std
```

```python
# now the preprocessing is complete 
df.describe()
# mean should be 0 and std should be 1
```
